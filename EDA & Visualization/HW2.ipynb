```python
new_df = df[['Age', 'Sex', 'Job', 'Housing', 'Saving accounts', 'Checking account', 'Credit amount', 'Duration', 'Purpose' ]]
new_df.head(10)
```

<div dir=rtl>
    ۱۷.(۶ نمره)
    یکی از مواردی که ممکن است تاثیر مخربی روی تحلیل ما بگذارد تفاوت اندازه‌های اعداد ستون‌های مختلف است(چرا؟ خیلی کوتاه توضیح دهید) برای همین بهتر است داده‌ها را نرمالایز کنیم.
    نرمالسازی داده در pca بسیار مهم است به دلیل اینکه pca سعی میکند که واریانس را ماکسیمم کند
در واقع pca داده ها را در جهتی قرار میدهد که واریانس مپ شدن داده ها به آن جهت بیشینه شود.
    بنابراین اندازه ی واریانس ماتریس کواریانس داده ها در تصمیم گیری جهت انتخابی تاثیر دارد و برای مقایسه ی منصفانه بین اهمیت ستون های مختلف حتما باید نرمالسازی صورت بگیرد.
</div>

```python
df_norm = (new_df - new_df.mean()) / (new_df.max() - new_df.min())
df_norm.head(10)
```

<div dir=rtl>
    ۱۸.(۳ نمره)
    حال نیاز است که ماتریس کواریانس این داده‌ها بدست بیاورید.
</div>

```python
data = np.array(df_norm)
cov_mat = np.cov(data.T)
print(cov_mat)
```

    ۱۹.(۲ نمره)

```python
fig = plt.figure(figsize=(14,11))
sns.heatmap(cov_mat, annot=True);
plt.show()
```

<div dir=rtl>
    ۲۰.(۴ نمره)
مقدار و بردار ویژه‌ی ماتریس کواریانس را بدست بیاورید.
</div>

```python
values, vectors = np.linalg.eig(cov_mat)
print('Eigen values: \n%s' %values)
print('Eigen vectors: \n%s' %vectors)
```

<div dir=rtl>
    ۲۱.(۳ نمره)
    برای انتخاب اجزای اساسی(Principle Components) سه کار ریز را انجام دهید:
    <br\>
هر بردار ویژه را با مقدار ویژه متناظرش مربوط کنید (می‌توانید بصورت جفتیpair آن‌ها را کنار هم قرار دهید)
        <br\>
            لیستی که بردار و مقدار ویژه‌ها دارید را بر اساس مقدار ویژه‌ها مرتب کنید
            <br\>
                مقدار ویژه‌های مرتب شده را به ترتیب نزولی در خروجی نمایش دهید
</div>

```python
from operator import itemgetter
# first task
eigen = []
for i in range(0, len(values)):
    eigen.append([values[i], vectors[i], df_norm.columns[i]])
# second task
eigen = sorted(eigen, key=itemgetter(0), reverse=True)
# third task
eigenvalues = []
eigenvectors = []
columns = []
for i in range(0, len(eigen)):
    eigenvalues.append(eigen[i][0])
    eigenvectors.append(eigen[i][1])
    columns.append(eigen[i][2])
print(eigenvalues)
# print(eigenvectors)
```

<div dir=rtl>
    ۲۲.(۴ نمره)
همه‌ی
    explained variance 
    را بدست بیاورید و در نمودار pie نمایش دهید
    حتما درصد هرکدام از اسلایس‌ها در نمودار مشخص شده‌باشد.
</div>

```python
variance_explained = []
for i in eigenvalues:
     variance_explained.append((i/sum(eigenvalues))*100)
print(variance_explained)
fig1, ax1 = plt.subplots()
ax1.pie(variance_explained, labels=columns, autopct='%0.1f%%', startangle=30)
ax1.axis('equal')
plt.show()
```

    ۲۵.(۵ نمره)
    در این قسمت می‌خواهیم مقایسه‌ای بین tSNE و PCA داشت باشیم.
    برای tSNE از scikit learn استفاده کنید و نیازی نیست که آنرا خودتان پیاده‌سازی کنید.
    تبدیل X را با استفاده از tSNE با دو مولفه اصلی بدست بیاورید.
    توجه کنید که tSNE پارامترهای مختلفی دارد که می‌توان با تنظیم آن‌ها به نتایج مختلفی رسید. در مورد انتخابتان برای این پارامترها توضیح دهید.
    از جمله پارامترهای tsne میتوان به موارد زیر اشاره کرد:
    1) n_components: در این جا چون میخواهیم داده ها را به فضای دوبعدی ببریم این پارامتر باید برابر با 2 باشد.
    2) perplexity: این پارامتر در واقع تعداد همسایه هایی را که باید برای هر نقطه در نظر گرفته شود نشان میدهد که معمولا بین 5 تا 50 تغییر میکند و هر چه تعداد داده ها بیشتر باشد این پارامتر هم بیشتر میشود.
    در این جا چون داده ها 1000 تا هستند یعنی اندازه ی متوسطی داند این پارامتر را برابر با 25 قرار میدهیم.
    3) early_exaggeration: این پارامتر درواقع نشان دهنده ی این است که دسته های مختلف داده ها در فضای با بعد کمتر چقدر از هم فاصله داشته باشند. مقدار پیشفرض این پارامتر 12 است. از آنجایی که این پارامتر چندان تاثیری ندارد همین پیشفرض مناسب است.
    4) learning_rate: ابن پارامتر درواقع میزان سرعت  gradient_descent را در پیشروی به سمت نقطه ی مینیموم نشان میدهد.
    اگر خیلی کم باشد سرعت کم بوده و اگر زیاد باشد ممکن است در local minimum گیر کند. پیشفرض 200 برای کار ما مناسب است.
    5) n_iter: بیشترین تعداد مرحله را برای بهینه سازی نشان میدهد. پیشفرض 1000 برای کار ما مناسب است.
    6) min_grad_norm: اگر نرم گرادیان از این پارامتر کم تر باشد عملیات بهینه سازی متوقف میشود. مقدار پیشفرض 0.0000001 مناسب است.
    7) metric: روش تعیین فاصله بین دو نقطه در فضای کاهش یافته را مشخص میکند. میتواند str یا اقلیدسی یا موارد دیگر باشد. در این جا همانطور که سر کلاس نیز گفته شد از اقلیدسی استفاده میکنیم چون نقاط بسیار نزدیک هم هستند جمع شدن فواصل اقلیدسی نقاط نزدیک هم در واقع همان فاصله دو نقطه دور از هم روی یک خم را تخمین میزند.
</div>
